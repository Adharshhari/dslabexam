```python
data=pd.read_csv("/content/dataset.csv")
```

```python
data
```

```python
plt.scatter(data['BM1'],data['Age'])
plt.xlabel('BM1')
plt.ylabel('Age')
```

```python
x=data.drop('Sugar',axis=1)
x
```

```python
y=data['Sugar']
```

```python
sc=StandardScaler()
```

```python
x=sc.fit_transform(x)
x
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x,y)
```

```python
xtest=[[43.6,40]]
xtrain=sc.fit_transform(xtest)
pred_value=knn.predict(xtrain)
```
